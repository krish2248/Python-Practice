```python
#encapsulation

class student:
    def __init__(self):
        self.__name  = "Krish"

    def getname(self):
        return self.__name
    
    def setname(self,name):
        self.__name = name

obj = student()

obj.setname("Khalnayak")

name = obj.getname()

print(name)
```
